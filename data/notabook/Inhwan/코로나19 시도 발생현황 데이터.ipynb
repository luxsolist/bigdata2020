```python
response = requests.get(url + queryParams)
response_text = response.text.encode('utf-8')
soup = BeautifulSoup(response_text, 'html.parser')
```

```python
items = soup.find_all('item')
```

```python
len(items)
```

```python
createdt_list = []
deathcnt_list = []
defcnt_list = []
gubun_list = []
incdec_list = []
isolclearcnt_list = []
isolingcnt_list = []
localocccnt_list = []
overflowcnt_list = []
qurrate_list = []
```

```python
len(items)
```

```python
for i in range(0, len(items)):
    # 등록일시분초
    createdt = items[i].select('createdt')
    if len(createdt) == 0 :
        createdt = None
        createdt_list.append(createdt)
    else :
        createdt = createdt[0].text
        createdt_list.append(createdt)
    
    # 사망자수
    deathcnt = items[i].select('deathcnt')
    if len(deathcnt) == 0 :
        deathcnt = None
        deathcnt_list.append(deathcnt)
    else :
        deathcnt = deathcnt[0].text
        deathcnt_list.append(deathcnt)
    
    # 확진자 수 
    defcnt = items[i].select('defcnt')
    if len(defcnt) == 0 :
        defcnt = None
        defcnt_list.append(defcnt)
    else :
        defcnt = defcnt[0].text
        defcnt_list.append(defcnt)
    
    # 시도명 
    gubun = items[i].select('gubun')
    if len(gubun) == 0 :
        gubun = None
        gubun_list.append(gubun)
    else :
        gubun = gubun[0].text
        gubun_list.append(gubun)
        
    # 전일대비 증감수 
    incdec = items[i].select('incdec')
    if len(incdec) == 0 :
        incdec = None
        incdec_list.append(incdec)
    else :
        incdec = incdec[0].text
        incdec_list.append(incdec)
    
    # 격리 해제 수
    isolclearcnt = items[i].select('isolclearcnt')
    if len(isolclearcnt) == 0 :
        isolclearcnt = None
        isolclearcnt_list.append(isolclearcnt)
    else :
        isolclearcnt = isolclearcnt[0].text
        isolclearcnt_list.append(isolclearcnt)
    
    # 격리중 환자수
    isolingcnt = items[i].select('isolingcnt')
    if len(isolingcnt) == 0 :
        isolingcnt = None
        isolingcnt_list.append(isolingcnt)
    else :
        isolingcnt = isolingcnt[0].text
        isolingcnt_list.append(isolingcnt)
    
    # 지역발생수
    localocccnt = items[i].select('localocccnt')
    if len(localocccnt) == 0 :
        localocccnt = None
        localocccnt_list.append(localocccnt)
    else :
        localocccnt = localocccnt[0].text
        localocccnt_list.append(localocccnt)
        
    # 해외유입수
    overflowcnt = items[i].select('overflowcnt')
    if len(overflowcnt) == 0 :
        overflowcnt = None
        overflowcnt_list.append(overflowcnt)
    else :
        overflowcnt = overflowcnt[0].text
        overflowcnt_list.append(overflowcnt)
    
    # 해외유입수
    qurrate = items[i].select('qurrate')
    if qurrate[0].text == '' or qurrate[0].text == '-' :
        qurrate = None
        qurrate_list.append(qurrate)
    else :
        qurrate = qurrate[0].text
        qurrate_list.append(qurrate)
```

```python
all_data = pd.DataFrame({'createdt' : createdt_list, 'deathcnt' : deathcnt_list , 'defcnt' : defcnt_list, \
                                  'gubun' : gubun_list, 'incdec' : incdec_list, 'isolclearcnt' : isolclearcnt_list, \
                                  'isolingcnt' : isolingcnt_list, 'localocccnt' : localocccnt_list, 'overflowcnt' : overflowcnt_list, \
                                  'qurrate' : qurrate_list})
```

```python
def areacode(gubun) : 
    if gubun == '서울' :
        return '1'
    elif gubun == '인천' :
        return '2'
    elif gubun == '대전' :
        return '3'
    elif gubun == '대구' :
        return '4'
    elif gubun == '광주' :
        return '5'
    elif gubun == '부산' :
        return '6'
    elif gubun == '울산' :
        return '7'
    elif gubun == '세종' :
        return '8'
    elif gubun == '경기' :
        return '31'
    elif gubun == '강원' :
        return '32'
    elif gubun == '충북' :
        return '33'
    elif gubun == '충남' :
        return '34'
    elif gubun == '경북' :
        return '35'
    elif gubun == '경남' :
        return '36'
    elif gubun == '전북' :
        return '37'
    elif gubun == '전남' :
        return '38'
    elif gubun == '제주' :
        return '39'
```

```python
covid_data['areacode'] = covid_data.apply(lambda x : areacode(x['gubun']), axis = 1)
```
