```python
import pandas as pd

df = pd.read_csv('C:/rsource2/감성분석/sentiment_matching.csv')
```

```python
df.head(30)
```

```python
cntTemp = df.groupby(['tour_id'], as_index=False).count()
```

```python
cntTemp.iloc[:,0:2]
```

```python
cnt = cntTemp.iloc[:,0:2]
```

```python
com_cnt = 0
tour_ids = []
count = []
temp_result =[]
```

```python
for i in range(len(cnt)):
    tempStr = []
    tempList = []
    
    com_next = com_cnt + cnt['r_word'][i]
    
    if com_next >= len(df):
        com_next = len(df)

    for j in range(com_cnt, com_next):
        tempStr = [df['r_word'][j],df['s_word'][j]]

        tempList.append(tempStr)
    
    tempList_sort = sorted(tempList, key=lambda x: (-x[1],x[0]))
    tour_ids.append(cnt['tour_id'][i])
    temp_result.append(str(tempList_sort))
    count.append(cnt['r_word'][i])

    com_cnt = com_cnt + cnt['r_word'][i]
```

```python
result = pd.DataFrame({'tour_id':tour_ids,'senti':temp_result,'senti_count':count})
```
