```python
from haversine import haversine
from sklearn.preprocessing import MinMaxScaler

# cur_location = (37.4798, 126.8828) # 우림라이온스빌딩B동
cur_location = (35.1791, 129.0751) # 부산광역시청
weight_dist = 1.0
weight_rcnt = 0.8

# gps 좌표로 거리(km) 계산
df['dist'] = df.apply(lambda x: haversine(cur_location, (x['mapy'], x['mapx'])), axis=1)

# 0~1 사이 값으로 normalization
scaler = MinMaxScaler()
df[['dist_scaled', 'readcount_scaled']] = scaler.fit_transform(df[['dist', 'readcount']])

# weight 계산
df['dist_scaled'] = weight_dist * (1 - df['dist_scaled'])
df['readcount_scaled'] = weight_rcnt * df['readcount_scaled']

# feature 합산
df['rank'] = df['dist_scaled'] + df['readcount_scaled']
```

```python
df.sort_values(by=['dist_scaled'], ascending=False).head(100).T
```
